```python
import struct

n = 632

with open("Result/J/spr_simulation_N_632_phi_1_a_2_U0_3e-20_k_0.bin", "br") as bf:
    print(bf)
    data = np.fromfile(bf, np.float64)
    print(data)
```

```python
t = 3750

def get_Pos_Vel(data, n, t):
    a = (t - 1)*(4*n + 1)
    b = t*(4*n + 1)
    
    x = []
    y = []
    
    V_x = []
    V_y = []
    
    i = 0
    for d in range(a,b - 1):
        if (b - a - i)%4 == 0:
            x.append(data[d])
        if (b - a - i + 1)%4 == 0:
            y.append(data[d])
        
        if (b - a - i + 2)%4 == 0:
            V_x.append(data[d])
        if (b - a - i + 3)%4 == 0:
            V_y.append(data[d])
        i = i + 1
        
    
    #print("X: ", x[:5])
    #print("Y: ", y[:5])
    
    #print("Velocity x: ", V_x[:5])
    #print("Velocity y: ", V_y[:5])
    
    #print("Lenght x: ", len(x))
    #print("Lenght y: ", len(y))    
    #print("Lenght V_x: ", len(V_x))
    #print("Lenght V_y: ", len(V_y))
    
    return [x, y, V_x, V_y]
```

# J

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

def f(t):  
    our_x, our_y, our_Vx, our_Vy =  get_Pos_Vel(data, n, t)
    h =  5e-07 * 2    # height
    w = 5e-07         # weight

    ells = []

    for i in range(0,len(our_x)):
        angle = np.arctan(our_Vx[i]/our_Vy[i])

        ells.append(Ellipse(xy=[our_x[i], our_y[i]],
                    width=w, height=h,
                    angle = angle*180/3.14))


    fig, ax = plt.subplots(subplot_kw={'aspect': 'equal'})
    for e in ells:
        ax.add_artist(e)
        e.set_clip_box(ax.bbox)
        e.set_alpha(1)
        e.set_facecolor([0,1,0])

    ax.set_xlim(0, 1.75e-05)
    ax.set_ylim(0, 1.75e-05)


    plt.gcf().set_size_inches(10,10)
    plt.title("S a-10 ..... t =" + str(t))

interactive_plot = interactive(f, t=(1, 7500, 20))
output = interactive_plot.children[-1]
output.layout.height = '780px'
interactive_plot
```
